Realiza una simulacion y grafica la soluci√≥n para un orificio de ùê¥‚Ñé=.5ùëö

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

%matplotlib inline
def tanque(Ah, Aw, h0):
    g = 9.81
    def velocidad(h,t):
        return -(Ah/Aw)*np.sqrt(2*g*h)
    t = np.linspace(0,6.5)
    h_t = odeint(velocidad,h0,t)
    
    plt.plot(t,h_t)
    plt.xlabel('time (min)')
    plt.ylabel('height (mts)')
    plt.grid()
    plt.show()
tanque(1,5,10)
print("Visualizando la gr√°fica con las condiciones dadas podemos deducir que la altura llegar√° a 2 mt. cuando el tiempo sea aproximadamente 4 min")
```

(10 puntos) ¬øQue pasa con una condici√≥n inicial menor (s(0)=21) ? ¬øY con una condici√≥n mayor a la capacidad de soporte?

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
def poblacion(s0,title):
    def crecimiento(s,t):
        return 0.05*s*((s/20)-1)*(1-(s/100))
    t = np.linspace(0,200)
    s_t = odeint(crecimiento,s0,t)
    
    plt.plot(t,s_t)
    plt.xlabel('time (min)')
    plt.ylabel('poblacion')
    plt.title(title)
    plt.grid()
    plt.show()
poblacion(21,'Grafica con s0=21')
print("La poblaci√≥n alcanza su m√°xima capacidad aproximadamente a los 105 min, respuesta tomada de la gr√°fica")
```

Problema 6 (estabilizar el sistema) Dise√±a una funci√≥n ùë¢ diferente de 0, para que el siguiente sistema no crezca, y se aproxime a 0 cuando t -> ‚àû.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline

def ds(s0):
    def cambio(s,t):
        u=-0.027*s*np.sqrt(s*t*0.5)
        return 0.06*u
    t = np.linspace(0,800)
    s_t = odeint(cambio,s0,t)
    
    plt.plot(t,s_t)
    plt.title('Simulaci√≥n del sistema')
    plt.show()
    
ds(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.integrate import odeint
%matplotlib inline

def dx(x0):
    def cambio(x,t):
        u=-0.005*np.exp(x*t*5.5)*np.sin(t/35)
        return 2*x*np.exp(2)+0.06*u
    t = np.linspace(0,800)
    x_t = odeint(cambio,x0,t)
    xref=0.5*np.sin(t/80)
    
    plt.plot(t,x_t)
    plt.plot(t,xref)
    plt.title('Simulaci√≥n del sistema')
    plt.show()
    
dx(10)
```
